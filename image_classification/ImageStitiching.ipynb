# Image stitching
allows you to stitch multiple images into a single image.
Example 1 shows the Python code for image stitching. This allows you to
specify multiple image files and stitch them together.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

img1 = cv.imread(r"D:\courses_by_mud\AI_Coding_ML_DL\image_processing\nature.jpg")
img2 = cv.imread(r"D:\courses_by_mud\AI_Coding_ML_DL\image_processing\R.jpeg")

if img1 is None or img2 is None:
    raise ValueError("One of the images could not be read.")

# Resize to same height
h1, w1 = img1.shape[:2]
h2, w2 = img2.shape[:2]
target_h = min(h1, h2)

img1r = cv.resize(img1, (int(w1 * target_h / h1), target_h))
img2r = cv.resize(img2, (int(w2 * target_h / h2), target_h))

combined = np.hstack([img1r, img2r])

# Save + show
out_path = r"D:\courses_by_mud\AI_Coding_ML_DL\image_processing\combined.jpg"
cv.imwrite(out_path, combined)
print("Saved:", out_path)

plt.figure(figsize=(14, 6))
plt.imshow(cv.cvtColor(combined, cv.COLOR_BGR2RGB))
plt.axis("off")
plt.title("Combined (Side-by-side)")
plt.show()
```
